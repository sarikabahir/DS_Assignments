```python
# converting into numpy array
wine_ary = wine.values
wine_ary
```

```python
# Normalizing the numerical data
wine_normal = scale(wine_ary)
wine_normal
```

```python
# Applying PCA fit transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
pca_values[:,4:]
```

```python
# the amount of variance that each PCA explains is
pca.explained_variance_
```

```python
# the amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# cumulative variance
var1 = np.cumsum(np.round(var, decimals = 4)*100)
var1
```

```python
# PCA Components matrix or convariance matrix
pca.components_
```

```python
wine.columns
```

```python
# Variance plot for PCA components obtained
plt.plot(var1, color = "blue");
```

```python
## Hierarchical Clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# create dendrogram
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_normal,'complete'))
```

```python
# Create Clusters
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(wine_normal),columns=['clustersid'])
y['clustersid'].value_counts()

```

```python
# adding clusters to dataset
wine2 = wine.copy()
wine2['clusterid'] = hc.labels_
wine2
```

```python
#K-Means Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range (1,6):
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
## Build cluster algorithm using k-3
```

```python
# cluster algorithm using k-3
cluster3 = KMeans(3, random_state=30).fit(wine_normal)
cluster3
```

```python
cluster3.labels_
```

```python
# assign clusters to the data set
wine3 = wine.copy()
wine3['cluster3id'] = cluster3.labels_
wine3
```
